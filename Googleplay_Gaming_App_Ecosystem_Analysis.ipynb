```python
N = len(totalgamecategory)     # Number of boxes

# generate an array of rainbow colors by fixing the saturation and lightness of the HSL representation of colour 
# and marching around the hue. 

c = ['hsl('+str(h)+',50%'+',50%)' for h in np.linspace(0, 360, N)]

#loading the data with filters
data = [{
    'y': gamedf.starrating[gamedf.gamecategory == j],
    'name': j,
    'type':'box',
    'marker':{'color': c[i]}
    } for i, j in enumerate(totalgamecategory)]


# format the layout
layout = {'xaxis': {'showgrid':False,'zeroline':False, 'tickangle':60,'showticklabels':False},
          'yaxis': {'zeroline':False,'gridcolor':'white'},
          'paper_bgcolor': 'rgb(233,233,233)',
          'plot_bgcolor': 'rgb(233,233,233)',
          }

py.offline.iplot(data)

```

```python
gamedf.inappprice = gamedf.inappprice.astype(str)
gamedf.appprice = gamedf.appprice.astype(str)
```

```python
def paymentmethod(inappprice, appprice):
    if appprice == 'nan' and inappprice == 'nan':
        return 'Free'
    elif appprice == 'nan' and inappprice != 'nan':
        return 'Free with In-App Purchases'
    elif appprice != 'nan' and inappprice == 'nan':
        return 'Paid'
    elif appprice != 'nan' and inappprice != 'nan':
        return 'Paid with In-App Purchases'
```

```python
#create a new column 'paymentmethod' to separate paid and free games
gamedf['paymentmethod'] = list(map(paymentmethod, gamedf.inappprice, gamedf.appprice))
gamedf
```

```python
#group by payment method and game category
ratingbypayment = gamedf.groupby(['paymentmethod', 'gamecategory'], as_index = False).agg(['mean'])['starrating']
freegames = ratingbypayment.loc['Free']
freewithinapp = ratingbypayment.loc['Free with In-App Purchases']
paidgames = ratingbypayment.loc['Paid']
paidwithinapp = ratingbypayment.loc['Paid with In-App Purchases']

```

```python
#barplots for ratings between free and paid games

free_games = go.Bar(x=freegames.index.tolist(),
                  y=freegames['mean'].values.tolist(),
                  name='Free Games',
                  marker=dict(color='rgb(166,206,227)'))

free_with_inapp = go.Bar(x=freewithinapp.index.tolist(),
                  y=freewithinapp['mean'].values.tolist(),
                  name='Free_with_Inapp',
                  marker=dict(color= 'rgb(31,120,180)'))


paid_games = go.Bar(x=paidgames.index.tolist(),
                y=paidgames['mean'].values.tolist(),
                name='Paid Games',
                marker=dict(color='rgb(178,223,138)'))


paid_with_inapp = go.Bar(x=paidwithinapp.index.tolist(),
                y=paidwithinapp['mean'].values.tolist(),
                name='Paid_with_Inapp',
                marker=dict(color= 'rgb(51,160,44)'))



data = [free_games, free_with_inapp, paid_games, paid_with_inapp]

layout = go.Layout(title="Average Rating for Paid and Free Games by Categories",
                xaxis=dict(title='Game Category'),
                yaxis=dict(title='Average Rating', range = [3.5,4.5]))

fig = go.Figure(data=data, layout=layout)

py.offline.iplot(fig)


```
