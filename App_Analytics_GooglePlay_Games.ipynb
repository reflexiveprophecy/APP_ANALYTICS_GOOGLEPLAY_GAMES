## RACING GAMES COMPARED TO OTHER GAMES

```python
#download latest plotly library
import plotly as py
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True)
```

```python
#the mean of ratings for every game category
gamegroup = gamedf.groupby(['gamecategory'])
gamegroup.agg(['mean'])['starrating']
```

```python
totalgamecategory = pd.unique(gamedf['gamecategory'])
totalgamecategory = totalgamecategory.tolist()
totalgamecategory 
```

```python
N = len(totalgamecategory)     # Number of boxes

# generate an array of rainbow colors by fixing the saturation and lightness of the HSL representation of colour 
# and marching around the hue. 

c = ['hsl('+str(h)+',50%'+',50%)' for h in np.linspace(0, 360, N)]

#loading the data with filters
data = [{
    'y': gamedf.starrating[gamedf.gamecategory == j],
    'name': j,
    'type':'box',
    'marker':{'color': c[i]}
    } for i, j in enumerate(totalgamecategory)]


# format the layout
layout = go.Layout(title="Average Rating for Every Game Category",
                xaxis=dict(title='Game Category'),
                yaxis=dict(title='Average Rating'))


fig = go.Figure(data=data, layout=layout)

py.offline.iplot(fig)

```

```python
gamedf.inappprice = gamedf.inappprice.astype(str)
gamedf.appprice = gamedf.appprice.astype(str)
```

```python
def paymentmethod(inappprice, appprice):
    if appprice == 'nan' and inappprice == 'nan':
        return 'Free'
    elif appprice == 'nan' and inappprice != 'nan':
        return 'Free with In-App Purchases'
    elif appprice != 'nan' and inappprice == 'nan':
        return 'Paid'
    elif appprice != 'nan' and inappprice != 'nan':
        return 'Paid with In-App Purchases'
```

```python
#create a new column 'paymentmethod' to separate paid and free games
gamedf['paymentmethod'] = list(map(paymentmethod, gamedf.inappprice, gamedf.appprice))
gamedf
```

```python
#group by payment method and game category
ratingbypayment = gamedf.groupby(['paymentmethod', 'gamecategory'], as_index = False).agg(['mean'])['starrating']
freegames = ratingbypayment.loc['Free']
freewithinapp = ratingbypayment.loc['Free with In-App Purchases']
paidgames = ratingbypayment.loc['Paid']
paidwithinapp = ratingbypayment.loc['Paid with In-App Purchases']

```

```python
#barplots for ratings between free and paid games

free_games = go.Bar(x=freegames.index.tolist(),
                  y=freegames['mean'].values.tolist(),
                  name='Free Games',
                  marker=dict(color='rgb(166,206,227)'))

free_with_inapp = go.Bar(x=freewithinapp.index.tolist(),
                  y=freewithinapp['mean'].values.tolist(),
                  name='Free_with_Inapp',
                  marker=dict(color= 'rgb(31,120,180)'))


paid_games = go.Bar(x=paidgames.index.tolist(),
                y=paidgames['mean'].values.tolist(),
                name='Paid Games',
                marker=dict(color='rgb(178,223,138)'))


paid_with_inapp = go.Bar(x=paidwithinapp.index.tolist(),
                y=paidwithinapp['mean'].values.tolist(),
                name='Paid_with_Inapp',
                marker=dict(color= 'rgb(51,160,44)'))



data = [free_games, free_with_inapp, paid_games, paid_with_inapp]

layout = go.Layout(title="Average Rating for Paid and Free Games by Categories",
                xaxis=dict(title='Game Category'),
                yaxis=dict(title='Average Rating', range = [3.5,4.5]))

fig = go.Figure(data=data, layout=layout)

py.offline.iplot(fig)


```

## WHY ARE RACING GAME RATED MORE POORLY THAN OTHERS?







```python
#Find the top 20 racing games
racinggames = gamedf[gamedf.gamecategory == 'Racing']
racinggames.ratingnums = list(map(pd.to_numeric,racinggames.ratingnums.str.replace(',','')))
racinggroup = racinggames.groupby(['appname']).agg(['mean'])[['ratingnums', 'starrating']]
topracinggames = racinggroup.sort_values(('starrating','mean'), ascending = False)[0:20]
topracinggames['ratingnums'] = topracinggames['ratingnums'].astype(int)
```

```python
#Find the bottom 50 racing games
botracinggames = racinggroup.sort_values(('starrating','mean'), ascending = True)[0:50]
botracinggames['ratingnums'] = botracinggames['ratingnums'].astype(int)


```

```python
#Scatterplot to show top racing game and bottom racing game ratings
trace0 = go.Scatter(
    x = topracinggames[('starrating','mean')].values.tolist(),
    y = topracinggames[('ratingnums','mean')].values.tolist(),
    name = 'Top 20 Racing Games',
    mode = 'markers',
    marker = dict(
        size = 20,
        color = 'rgba(152, 10, 100, .8)',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)

trace1 = go.Scatter(
    x = botracinggames[('starrating','mean')].values.tolist(),
    y = botracinggames[('ratingnums','mean')].values.tolist(),
    name = 'Worst 50 Racing Games',
    mode = 'markers',
    marker = dict(
        size = 20,
        color = 'rgba(255, 150, 193, .9)',
        line = dict(
            width = 2,
        )
    )
)

data = [trace0, trace1]

layout = dict(title = 'TOP AND WORST RACING GAMES BY RATINGS AND RATING COUNTS',
              xaxis=dict(title='Ratings'),
                yaxis=dict(title='Rating Counts')
             )

fig = dict(data=data, layout=layout)
py.offline.iplot(fig, filename='Top and Worst Racing Games by Ratings and Number of Ratings')
```
